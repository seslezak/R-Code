```python
top_contributors = merged.fillna('').groupby(
    ['contributor_lastname', 'contributor_firstname', 'committee_position']
).amount.sum().reset_index().sort_values('amount', ascending = False)

top_contributors.head()
```

```python
alt.Chart(top_contributors.head(10)).mark_bar().encode(
    y = alt.Y('contributor_fullname:N', sort = '-x', title = None),
    x = alt.X('amount:Q', title = "Contribution Amount"),
    color = alt.Color('committee_position:N', title = 'Postiion'),
    tooltip = ['contributor_fullname:N', 'amount:Q', 'committee_position:N']
).properties(
    title = 'Top Contributors, CalProp 64, in 2016',
    width = 500, height = 400
)
```

```python
top_contributors["contributor_fullname"] = top_contributors["contributor_firstname"] + " " + top_contributors["contributor_lastname"]
```
